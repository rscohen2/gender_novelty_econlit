```python
import pandas as pd

df = pd.read_csv('df_litecon.csv')
```

```python
"""
Perform OLS to see effect of domain and gender
"""
```

```python
import statsmodels.formula.api as smf

# Citations model
model_cite = smf.ols(
    formula='citation_counts ~ C(gender) + year +  C(domain)',
    data=df
).fit()

print(model_cite.summary())

# Novelty model
model_nov = smf.ols(
    formula='novelty ~ C(gender) + year  + C(domain)',
    data=df
).fit()

print(model_nov.summary())

model_prec = smf.ols(
    formula='precocity ~ C(gender) + year + C(domain)',
    data=df
).fit()

print(model_prec.summary())
```

```python
import statsmodels.formula.api as smf

# Citations model
model_cite = smf.ols(
    formula='citation_counts ~ C(gender) + year + C(journal) + C(domain)',
    data=df
).fit()

print(model_cite.summary())

# Novelty model
model_nov = smf.ols(
    formula='novelty ~ C(gender) + year + C(journal) + C(domain)',
    data=df
).fit()

print(model_nov.summary())

model_prec = smf.ols(
    formula='precocity ~ C(gender) + year + C(journal) + C(domain)',
    data=df
).fit()

print(model_prec.summary())
```
